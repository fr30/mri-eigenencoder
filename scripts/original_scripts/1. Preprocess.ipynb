```python
import joblib
import torch

import numpy as np

# LOAD DATA: Put EEG, EMG, LABEL, SUBJECT Data in the folder ../EEG_EMG_DATA

eeg_db = joblib.load('../EEG_EMG_DATA/eeg_train_trial.pt')
emg_db = joblib.load('../EEG_EMG_DATA/emg_filtered_train_trial.pt')

label = joblib.load('../EEG_EMG_DATA/label_train_trial.pt')
subject = joblib.load('../EEG_EMG_DATA/sub_train_trial.pt')

emg_data = torch.from_numpy(emg_db).float()
eeg_data = torch.from_numpy(eeg_db).float()

emg_data = emg_data.permute(2, 0, 1)
eeg_data = eeg_data.permute(2, 0, 1)

index = np.arange(0, emg_data.shape[0])

# Normalization: Normalize signals by the magnitude of the first sample

emg_data = emg_data/emg_data[0].max()
eeg_data = eeg_data/eeg_data[0].max()

# Remove Outliers: Remove outlier samples that has magnigtude larger than 5 or smaller than -5

if_keep = (emg_data>5).sum(1).sum(1)
emg_data = emg_data[torch.where(if_keep<1)[0]]
eeg_data = eeg_data[torch.where(if_keep<1)[0]]
label = label[torch.where(if_keep<1)[0]]
subject = subject[torch.where(if_keep<1)[0]]
index = index[torch.where(if_keep<1)[0]]

if_keep = (emg_data<-5).sum(1).sum(1)
emg_data = emg_data[torch.where(if_keep<1)[0]]
eeg_data = eeg_data[torch.where(if_keep<1)[0]]
label = label[torch.where(if_keep<1)[0]]
subject = subject[torch.where(if_keep<1)[0]]
index = index[torch.where(if_keep<1)[0]]

if_keep = (eeg_data>5).sum(1).sum(1)
emg_data = emg_data[torch.where(if_keep<1)[0]]
eeg_data = eeg_data[torch.where(if_keep<1)[0]]
label = label[torch.where(if_keep<1)[0]]
subject = subject[torch.where(if_keep<1)[0]]
index = index[torch.where(if_keep<1)[0]]

if_keep = (eeg_data<-5).sum(1).sum(1)
emg_data = emg_data[torch.where(if_keep<1)[0]]
eeg_data = eeg_data[torch.where(if_keep<1)[0]]
label = label[torch.where(if_keep<1)[0]]
subject = subject[torch.where(if_keep<1)[0]]
index = index[torch.where(if_keep<1)[0]]

# Match labels: Merge different subtask movements of reaching, grasping and twisting into primary movements

new_label = []
for k in label:
    if k in range(1, 7):
        new_label.append(0) # reaching
    if k in range(7, 10):
        new_label.append(1) # grasping
    if k in range(10, 12):
        new_label.append(2) # twisting
        
new_label = torch.tensor(new_label).long().view(-1)

# Save processed signals for experiments

!mkdir './run_experiments_ssl_signal/'
np.save('./run_experiments_ssl_signal/subject.npy', subject)
np.save('./run_experiments_ssl_signal/label.npy', label)
np.save('./run_experiments_ssl_signal/new_label.npy', new_label)
np.save('./run_experiments_ssl_signal/eeg_data.npy', eeg_data)
np.save('./run_experiments_ssl_signal/emg_data.npy', emg_data)
```
